```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/datasets/imranalishahh/marketing-and-product-performance-dataset/marketing_and_product_performance.csv')

df.head()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Budget', y='Revenue_Generated')

plt.title("RQ1: Budget vs Revenue")

plt.savefig('/kaggle/working/RQ1.pdf', bbox_inches='tight')
plt.show()

rq1_table = df[['Budget', 'Revenue_Generated']].describe()
rq1_table.to_csv('/kaggle/working/RQ1.csv')
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Clicks', y='Conversions')

plt.title("RQ2: Clicks vs Conversions")

plt.savefig('/kaggle/working/RQ2.pdf', bbox_inches='tight')
plt.show()

rq2_table = df[['Clicks', 'Conversions']].describe()
rq2_table.to_csv('/kaggle/working/RQ2.csv')
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Discount_Level', y='Units_Sold')

plt.title("RQ3: Discount vs Units Sold")

plt.savefig('/kaggle/working/RQ3.pdf', bbox_inches='tight')
plt.show()

rq3_table = df.groupby('Discount_Level')['Units_Sold'].mean()
rq3_table.to_csv('/kaggle/working/RQ3.csv')
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Subscription_Tier', y='Customer_Satisfaction_Post_Refund')

plt.title("RQ4: Subscription vs Satisfaction")

plt.savefig('/kaggle/working/RQ4.pdf', bbox_inches='tight')
plt.show()

rq4_table = df.groupby('Subscription_Tier')['Customer_Satisfaction_Post_Refund'].mean()
rq4_table.to_csv('/kaggle/working/RQ4.csv')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['ROI'], bins=30, kde=True)

plt.title("RQ5: ROI Distribution")

plt.savefig('/kaggle/working/RQ5.pdf', bbox_inches='tight')
plt.show()

rq5_table = df[['Budget', 'Clicks', 'Conversions', 'ROI']].corr()
rq5_table.to_csv('/kaggle/working/RQ5.csv')
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Bundle_Price', y='Units_Sold')

plt.title("RQ6: Bundle Price vs Units Sold")

plt.savefig('/kaggle/working/RQ6.pdf', bbox_inches='tight')
plt.show()

rq6_table = df[['Bundle_Price', 'Units_Sold']].describe()
rq6_table.to_csv('/kaggle/working/RQ6.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

features = ['Budget', 'Clicks', 'Conversions', 'Units_Sold', 'Discount_Level']

X = df[features]
y = df['Revenue_Generated']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)

pred = model.predict(X_test)

print("MSE:", mean_squared_error(y_test, pred))

plt.figure(figsize=(8,5))
sns.scatterplot(x=y_test, y=pred)

plt.title("RQ7: Actual vs Predicted Revenue")

plt.savefig('/kaggle/working/RQ7.pdf', bbox_inches='tight')
plt.show()

pd.DataFrame({'Actual': y_test, 'Predicted': pred}).to_csv('/kaggle/working/RQ7.csv')
```
